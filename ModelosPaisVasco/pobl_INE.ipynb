```python
fldr = f'D:/Ctym/p/p189_ModelosPaisVasco/5_Trabajo/dev/outputs'

import pandas as pd
import os
import glob

file_pattern = os.path.join(fldr,'6****.csv')
dataframes = []

for file_path in glob.glob(file_pattern):
    df = pd.read_csv(file_path, sep = ';', encoding = 'utf-8')
    dataframes.append(df)

pobl = pd.concat(dataframes, ignore_index=True)
pobl.to_csv(os.path.join(fldr,'Pobl_munic_edad_sexo.csv'), encoding= 'utf-8', index=False)
```
